```python
class Rental_income():
    def __init__(self):
        self.income = 0
        self.total_expenses = 0
        self.cashflow = 0
        self.rent = 0

    def calc_income(self):
        income_dict = {}
        rent = float(input("How much are you charging for rent? "))
        if rent > 0:
            income_dict['rent'] = rent
        self.rent = rent
        laundry_fee = float(input("How much are laundry fees? "))
        if laundry_fee > 0:
            income_dict['laundry'] = laundry_fee
        storage_fee = float(input("How much are storage costs? "))
        if storage_fee > 0:
            income_dict['storage'] = storage_fee
        
        other =input("Do you take fees for anything else? yes/no")
        while other.lower() == 'yes':
            
            new_income = input("what else do you take fees for? ")
            if new_income:
                cost = float(input(f"How much do you charge for {new_income}"))
                if cost > 0:
                    income_dict[new_income] = cost
            other = input("Is there any other fees? ")

        self.income = sum(income_dict.values())
        print(f"Your total income is {self.income}")
        print(f"The income is from: ")
        for k,v in income_dict.items():
            print(f"{k} : {v}")


    def calc_expenses(self):
        expense_dict = {}

        tax_percent = float(input("What percent do you pay in taxes? "))
        
        tax = (tax_percent * self.rent ) / 100

        if tax > 0:
            expense_dict['taxes'] = tax

        insurance = float(input("How much does insurance cost on the property? "))
        if insurance > 0:
            expense_dict['insurance'] = insurance

        utilities = {}
        who_pays = input("Does the renter pay their own utilities? ").lower()
        if who_pays == 'no':
            electric = float(input("What do you pay for electricity on the property? "))
            if electric > 0:
                utilities['electric'] = electric
            water = float(input("What do you pay for water on the property?"))
            if water > 0:
                utilities['water'] = water
            sewage = float(input("What do you pay for sewage on the property? "))
            if sewage > 0:
                utilities['sewage'] = sewage
            garbage = float(input("What is the garbage fee you pay? "))
            if garbage > 0:
                utilities['garbage'] = garbage
            gas = float(input("What do you pay for gas fees on the property? "))
            if gas > 0:
                utilities['gas'] = gas
        
        utility_expense = sum(utilities.values())
        if utility_expense > 0:
            expense_dict['utilites'] = utility_expense
        print(f"The breakdown of cost for utilities is: ")
        for u,c in utilities.items():
            print(f"{u} : {c}")

        hoa_fees = float(input("What are the HOA fees on the property? "))
        if hoa_fees > 0:
            expense_dict['HOA'] = hoa_fees
        
        lawncare = float(input("What do you put into lawncare or snow each month? "))
        if lawncare > 0:
            expense_dict['Lawn/snow'] = lawncare
        
        vacancy = float(input("What percentage of your income goes to vacancy? "))
        if vacancy > 0:
            expense_dict['Vacancy'] = (vacancy * self.income) / 100
        
        repair = float(input('How much are you putting away for repairs? '))
        if repair > 0:
            expense_dict['Repair fund'] = repair

        capex = float(input("How much are you putting away for capital expenditure? "))
        if capex > 0:
            expense_dict['capital expenditure'] = capex

        prop_management = float(input('What percentage of the income goes to the property manager? '))
        if prop_management > 0:
            expense_dict['prop_management'] = (prop_management * self.rent) / 100

        mortgage = float(input('What is the mortgage payment on the property? '))
        if mortgage > 0:
            expense_dict['mortgage'] = mortgage

        self.total_expenses = sum(expense_dict.values()) 
        print(f"Your total monthly costs add up to {self.total_expenses}")
        print(f"The expenses are: ")

        for k,v in expense_dict.items():
            print(f"{k} : {v}")
        

    def calc_cashflow(self):
        self.cashflow = self.income - self.total_expenses
        print(f'Your cash flow is {self.cashflow}')

    def calc_ROI(self):
        invested = {}
        down_pay = float(input("What was the down payment on the property? "))
        if down_pay > 0:
            invested['Down payment'] = down_pay
        closing_cost= float(input("What were the closing costs of the property? "))
        if closing_cost > 0:
            invested['Closing costs'] = closing_cost
        repair_cost = float(input("What was the total cost of repairs on the property? "))
        if repair_cost > 0:
            invested['repair_cost'] = repair_cost
        misc_cost = float(input("What is the total of any other expenses? "))
        if misc_cost > 0:
            invested['misc costs'] = misc_cost

        annual_profit = self.cashflow * 12
        ROI = (annual_profit / sum(invested.values())) * 100
        print(f"Your rate of investment is {ROI}")


property1 = Rental_income()

property1.calc_income()
property1.calc_expenses()
property1.calc_cashflow()
property1.calc_ROI()

```
